```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from os import listdir
from natsort import natsorted
import sys
sys.path.append("..")
sys.path.append("../src")
sys.path.append("../utils")
sys.path.append("../phantoms")
sys.path.append("../projections")
from src.phantoms import create_phantoms
```

```python
phantoms = []
phantom_dir = "../phantoms/"
dir_names = natsorted(listdir(phantom_dir))
print("Phantom directories:",dir_names)
print()
for dir_n in dir_names:
    # take current family
    curr_phants = natsorted(listdir(phantom_dir+dir_n))
    print(curr_phants)
    
    for phant_name in curr_phants:
        print(phant_name)
        img_np = np.array(Image.open(phantom_dir+dir_n+"/"+phant_name))
        phantoms.append(img_np)
```
