```python
import numpy as np
from os import listdir
import matplotlib.pyplot as plt
from PIL import Image
import sys
sys.path.append("..")
sys.path.append("../src")
from src.utils import *

plt.style.use('seaborn-whitegrid')
```

```python
labels = [("DART_sart", "c" ), ("DART_sirt", "g"), ("SART", "r"), ("SIRT", "m"), ("DART_fbp", "b"), ("RBF", "y")]

n_projections = [2, 4, 6, 8, 10, 12, 14, 16, 18]
angle_range = [20, 40, 60, 80, 100, 120, 140, 160, 180]
tick_labels = ["("+str(n_projections[i])+","+str(angle_range[i])+")" for i in range(len(n_projections))]

parent_path = '../results/proj_ang_comb/'
save_dir = "../plots_for_report/"
```

```python
res_path = f"{parent_path}paws/"
title="Effect of angles and projections on Paw phantoms"
xlabel="Number of projections and angle range"
ylabel="Mean Absolute Pixel Error"
ylim=50
save_name= f"{save_dir}proj_ang_comb_paws.png"
# save_name = None
plot_results(res_dir=res_path, labels=labels,
            title=title, 
            tick_labels=tick_labels,
            xlabel=xlabel, x_rotate=-45,
            ylabel=ylabel, ylim=5,
            save_name=save_name)
```

```python
res_path = f"{parent_path}clouds/"
title="Effect of angles and projections on Cloud phantoms"

ylim=50
save_name= f"{save_dir}proj_ang_comb_clouds.png"

plot_results(res_dir=res_path, labels=labels,
            title=title, 
            tick_labels=tick_labels,
            xlabel=xlabel, x_rotate=-45,
            ylabel=ylabel, ylim=1.3,
            save_name=save_name)
```

```python
res_path = f"{parent_path}semilunars/"
title="Effect of angles and projections on Semilunar phantoms"

ylim=50
save_name= f"{save_dir}proj_ang_comb_semilunars.png"

plot_results(res_dir=res_path, labels=labels,
            title=title, 
            tick_labels=tick_labels,
            xlabel=xlabel, x_rotate=-45,
            ylabel=ylabel, ylim=3,
            save_name=save_name)
```
