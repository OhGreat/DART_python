```python
labels = [("DART_sart", "c" ), ("DART_sirt", "g"), ("SART", "r"), ("SIRT", "m"), ("DART_fbp", "b"), ("RBF", "y")]
tick_labels = [0.02, 0.04, 0.06, 0.08, 0.1, 0.12, 0.14]
tick_labels = [str(np.round(tick*100,2))+"%" for tick in tick_labels]

parent_path = '../results/gray_levels_rand/'
save_dir = "../plots_for_report/"
```
