```python
labels = ["DART_fbp", "DART_sart", "DART_sirt", "FBP", "SART", "SIRT"]
tick_labels = [ 0, 1500, 3000, 4500, 6000, 7500, 9000]
tick_labels = [str(np.round((noise/53418)*100, 1))+"%" for noise in tick_labels]
```
